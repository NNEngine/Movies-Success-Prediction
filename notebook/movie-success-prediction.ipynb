```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

import pickle
```

```python
df = pd.read_csv("/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv")
```

```python
df.head()
```

```python
df.drop(columns = ["Poster_Link", "Overview"], inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
obj_col_list = ["Series_Title", "Certificate", "Genre", "Director", "Star1","Star2", "Star2", "Star3", "Star4"]

encoder = LabelEncoder()

for col in obj_col_list:
    df[col] = encoder.fit_transform(df[col])
```

```python
df
```

```python
df["Runtime"] = df["Runtime"].str.replace(r'[^0-9]', "", regex = True)
df["Gross"] = df["Gross"].str.replace(r'[^0-9]', "", regex = True)
```

```python
df
```

```python
df.info()
```

```python
df["Released_Year"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df["Runtime"] = df["Runtime"].astype("int64")

df["Released_Year"] = df["Released_Year"].str.extract(r'(\d{4})')
df["Released_Year"] = df["Released_Year"].astype("int64")

df["Gross"] = df["Gross"].astype("int64")
```

```python
df["Gross"]
```

```python
df["Released_Year"].isnull().sum()
```

```python
df.info()
```

```python
plt.figure(figsize = (25, 8), dpi = 100, facecolor = "white", edgecolor = "black")
plt.boxplot(
    df
)
plt.title("Box Plot for Outlier Detection", fontsize = 20, fontweight = "bold", color = "black", loc = "center")
plt.show()
```

```python
scaler = StandardScaler()
df_array = scaler.fit_transform(df)
```

```python
df_array
```

```python
df
```

```python
df = pd.DataFrame(df_array, columns = df.columns)
```

```python
df
```

```python
X = df.drop(columns = "Gross")
Y = df["Gross"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
r2_score(Y_test, Y_pred)
```

```python
with open("movie_model.pkl", "wb") as f:
    pickle.dump(model, f)
```
